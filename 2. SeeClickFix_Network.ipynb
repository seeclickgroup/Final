# 2. Network Analysis

**The focus of the network analysis included trying to understand:**

**Who are the "power" users that have significant engagement across the platform?**  In order to conduct this analysis we linked user data with the issue history data to trace user interactions to specific issues.  This provided an understanding of how users were interacting with each other.  Additionally, we labeled the users as those who are normal users and those who are "official" users (i.e. can Acknowledge issues).  The application would be for a city official who is interested in how the community is engaging with the platform and is interested in knowing about sub-communities that frequently interact.

```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#import urllib
#import json
#import datetime as dt
from matplotlib import rcParams
#import scipy as sp
#from collections import defaultdict
import networkx as nx

#tell pandas to display wide tables as pretty HTML tables
pd.set_option('display.width', 500)
pd.set_option('display.max_columns', 100)

#colorbrewer2 Dark2 qualitative color table
dark2_colors = [(0.10588235294117647, 0.6196078431372549, 0.4666666666666667),
                (0.8509803921568627, 0.37254901960784315, 0.00784313725490196),
                (0.4588235294117647, 0.4392156862745098, 0.7019607843137254),
                (0.9058823529411765, 0.1607843137254902, 0.5411764705882353),
                (0.4, 0.6509803921568628, 0.11764705882352941),
                (0.9019607843137255, 0.6705882352941176, 0.00784313725490196),
                (0.6509803921568628, 0.4627450980392157, 0.11372549019607843)]

rcParams['figure.figsize'] = (10, 6)
rcParams['figure.dpi'] = 150
rcParams['axes.color_cycle'] = dark2_colors
rcParams['lines.linewidth'] = 2
rcParams['axes.facecolor'] = 'white'
rcParams['font.size'] = 14
rcParams['patch.edgecolor'] = 'white'
rcParams['patch.facecolor'] = dark2_colors[0]
rcParams['font.family'] = 'StixGeneral'

def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chartjunk by stripping out unnecesasry plot borders and axis ticks
    
    The top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
        
```

#### FILE READ

```python
#read Raw User Info DataFrame
user_df = pd.read_csv('/SCF_Users_ATL.csv')
```

```python
#read Raw Issue History DataFrame
issue_history_df = pd.read_csv('/SCF_Issue_History_ATL.csv')
```

#### END FILE READ

```python
#Link users to their interactions with issues
user_int_list = []
user_int_count_list = []

#look up user ids in the issue history dataframe and pass all issue ids that a given user id is attached to
for i in user_df['id'].index:
    user_int_count_list.append(len(issue_history_df.issue_id[issue_history_df['user_id'] == int(user_df['id'][i])].values))
    user_int_list.append(issue_history_df.issue_id[issue_history_df['user_id'] == int(user_df['id'][i])].values)

#place the issue ids and a count of total interactions into the user dataframe    
user_df['issue_interactions'] = user_int_list
user_df['issue_int_count'] = user_int_count_list
```

```python
#link issues to the users who interact on them
issue_users_list = []
issue_id_list = []
users_count_list = []

for issue in issue_history_df['issue_id'].unique():
    issue_users = issue_history_df[issue_history_df['issue_id'] == issue].user_id.unique()
    issue_users = issue_users[~np.isnan(issue_users)]
    user_count = len(issue_users)
    if len(issue_users) == 0:
            issue_users = np.nan
    users_count_list.append(user_count)
    issue_users_list.append(issue_users)
    issue_id_list.append(issue)

issue_df = pd.DataFrame(index=issue_id_list)
issue_df['issue_id'] = issue_id_list
issue_df['user_list'] = issue_users_list
issue_df['user_count'] = users_count_list
issue_df = issue_df[~issue_df.user_list.isnull()]
issue_df.reset_index(inplace=True)
```

```python
#create a network

g = nx.Graph()
for user in user_df.id:
    g.add_node(int(user))
    if user_df[user_df['id'] == user].can_ack == True:
        g.node[user]['color'] = 'b'
    else:
        g.node[user]['color'] = 'r'
    
for user_list in issue_df.user_list:
    if len(user_list) > 1:
        for user in user_list[1:]:
            if g.has_edge(int(user_list[0]), int(user)):
                g[user_list[0]][user]['weight'] += 1
            else:
                g.add_edge(int(user_list[0]), int(user), weight=1)
                
#set difference attribute
for edge in g.edges():
    g[edge[0]][edge[1]]['difference'] = 1./g[edge[0]][edge[1]]['weight']

g.remove_node(9846) #remove node in the issue_history dataframe
```

```python
#this makes sure draw_spring results are the same at each call
np.random.seed(1)  

color = [g.node[user]['color'] for user in g.nodes()]

#determine position of each node using a spring layout
pos = nx.spring_layout(g, iterations=200)

#plot the edges
nx.draw_networkx_edges(g, pos, alpha = .5)

#plot the nodes
nx.draw_networkx_nodes(g, pos, node_color=color)

#draw the labels
lbls = nx.draw_networkx_labels(g, pos, alpha=5, font_size=8)

#coordinate information is meaningless here, so let's remove it
plt.xticks([])
plt.yticks([])
remove_border(left=False, bottom=False)
```

This graph provides a basic understanding of the network, but in order to identify the "power" users we will determine the closeness centrality for each user.  Finally, it may be more effective to explore and visualize the network in Gephi.  Therefore we will export the network and explore.

```python
bet = nx.closeness_centrality(g, distance='difference')
connected = sorted(bet, key=lambda x: -bet[x])

print "Highest closeness"
for user in connected[:5]:
    print "%20.20s\t%0.3f" % (user, bet[user])
    
nl = g.nodes()
plt.figure(figsize=(15, 4))
x = np.arange(len(nl))
y = np.array([bet[n] for n in nl])
c = np.array([g.node[n]['color'] for n in nl])

ind = np.argsort(y)
y = y[ind]
c = c[ind]

plt.bar(x, y, color=c, align='center', width=.8)

remove_border(left=None, bottom=None)
ticks = plt.xticks(x, [nl[i] for i in x[ind]], 
                   rotation='vertical', fontsize=7)
limits = plt.xlim(-1, x[-1] + 1)
```

By calculating the closeness centrality of the users we can quickly understand the distribution of interaction among users.  The top five users, identified above, are the best representation of the "power" users who are most active on the platform and in the Atlanta SeeClickFix Community.

To visualize how exactly these users are connected and the frequency of interaction among users it is better to use Gephi for exploration.

**Conclusions**

The user community in Atlanta is comprised of 146 users.  Almost half of those users simply interact with the platform by reporting a small number of issues. The other half of users contribute in numerous ways including reporting, voting, commenting, and closing issues. Additionally, there is a small group of "power users" who are well connected across Atlanta users. The graph above displays the overall network, using user ids as the nodes and common issue interactions as the edges. The edges are weighted by the number of common issue interactions (comprising of commenting, voting, following, acknowledging, and closing). Therefore, thicker lines indicate more interaction among a pair of users.

On the above network graph, the circled users are considered the top "power users" using closeness centrality as the measure for influence or importance.  The above chart shows the closeness score for all of the users.  The top five user ids and scores are highlighted

In the context of a 311 platform, it is valuable to identify the most active users and communities.  This information can be useful in detecting fraudulent reports or recognition of active and valuable users / communities. In order to do this, the 311 platform must be structured to capture user information when they report issues. For example, when a user sends in an issue over the phone there should be an attempt to capture information that can then be linked to the same user when they submit online or via a mobile app.
