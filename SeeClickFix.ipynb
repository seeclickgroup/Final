### Sample Bias and Caveats

When considering the data presented and its applicability to an Atlanta 311 platform, it is valuable to note the potential biases in the sample. One major factor is that SeeClickFix is an online platform, therefore users must be able to access the internet in order to submit issues. There are disparities in internet access across the City of Atlanta; more information on broadband access in Atlanta can be found at the broadband.gov website.

Additionally, individual users and interest groups can bias the sample by focusing on specific issue types and reporting on them to the exclusion of others. This will bias the sample by over-inflating specific issues relative to other issues happening in Atlanta. City officials should consider this when responding to quantities and types of issues. Upon launch of the Atlanta 311 platform, it should be expected that similar types of groups and issue-driven individuals will use that platform to report their specific concerns. For instance, in our network analysis, we were able to identify specific users that had low centrality scores (i.e. low interaction with other users), but high submissions of issues, which may bias issue reports based on the predilections of highly motivated individuals. This type of activity should be further investigated.

```python
def month_graph_by_type(Type, df):

    # Limiting to certain type
    
    smalldf2 = df[df['type'] == Type]
    smalldf2 = smalldf2[smalldf2['type'] != 'FALSE']

    grouped = smalldf2.groupby(['year', 'month'])
    month_count_type = grouped['date'].count()
    
    
    dict2 = {'freq':month_count_type}
    df2 = pd.DataFrame(dict2)
    df2 = df2.reset_index()
    
    monthyear2 = []
    for i in range(len(df2)):
        temp = str(df2['month'][i]) + "/" + str(df2['year'][i])
        monthyear2.append(temp)
    df2['month_year'] = monthyear2
    
    x = [dt.datetime.strptime(d,'%m/%Y').date() for d in df2['month_year']]
    df2['date'] = x
    
    
    # Graph by month
    
    rcParams['figure.figsize'] = (15, 6)
    
    plt.xlabel('Date')
    plt.ylabel('Number of Issues per month')
    plt.title("Number of Issues Submitted per Month in Atlanta")
    x = df2['date']
    y = df2['freq']
    plt.plot(x,y, label = Type)
    plt.legend(loc='best')
    remove_border()
```

```python
# More manageable graph
interesting_list = ['Animal', 'Vacant/Delapidated', 'Overgrowth', 'Traffic', 'Pothole/paving', 'Signage', 'Tree', 'Water/sewage system', 'Street cleaning', 'Sidewalk', 'Street lighting', 'Garbage', 'Dumping/debris']
    
for each_type in interesting_list:
    month_graph_by_type(each_type, atl_df)
    
```
